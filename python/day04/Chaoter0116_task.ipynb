__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
inventory = {'팥붕어빵' : 10, '슈크림붕어빵' : 10, '초코붕어빵' : 10}
while True :
    print("\n1. 팥붕어빵 (1000원)")
    print("2. 슈크림붕어빵 (1200원)")
    print("3. 초코붕어빵 (1500원)")
    user_input = input("주문할 붕어빵의 번호를 입력하세요: ")
    order = {} #주문내역을 초기화 한다.
    if user_input == "1" : #1번을 입력했을경우 '팥붕어빵이 선택된다.
        kind = "팥붕어빵"
    elif user_input == "2" : #2번을 입력했을경우 '슈크림붕어빵이 선택된다.
        kind = "슈크림붕어빵"
    elif user_input == "3" : #3번을 입력했을경우 '초코붕어빵이 선택된다.
        kind = "초코붕어빵"
    else :
        print("주문을 잘 못 했습니다. 다시 주문해 주세요")
        continue
    
    quantity = int(input(f'{kind} 몇개를 주문 하시겠습니까?'))
    if kind in order :
        order[kind] += quantity
    else :
        order[kind] = quantity
    print("\n 주문내역 :", order)
    print("프로그램을 종료합니다.")
    break
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
inventory = {'팥붕어빵' : 10, '슈크림붕어빵' : 10, '초코붕어빵' : 10}

while True :
    print("\n 1. 팥붕어빵 : 1000원")
    print(" 2. 슈크림붕어빵 : 1200원")
    print(" 3. 초코붕어빵 : 1500원")
    print(" 4. 종료")
    user_input = input("주문하실 붕어빵 번호를 입력해 주새요:")
    order = {}
    if user_input == "1" :
        kind = "팥붕어빵"
    elif user_input == "2" :
        kind = "슈크림붕어빵"
    elif user_input == "3" :
        kind = "초코붕어빵"
    else :
        print("번호를 잘 못 입력하셨습니다. 다시 입력해 주세요")
        continue
    amount = int(input(f"{kind}를 선택하셨습니다. 수량을 입력해주세요:"))
    
    if kind in inventory :
        if inventory[kind] >= amount :
            inventory[kind] -= amount
            if kind in order :
                order[kind] += amount
            else : 
                order[kind] = amount
            print(f'{kind} {amount}개 판매완료')
        else :
            print(f"{kind}의 재고가 부족합니다. 최대 {inventory[kind]}만큼만 주문 가능합니다.")
            continue 
    else :
        print("판매하지 않는 상품입니다.")
        continue

    print("\n주문이 완료되었습니다.\n 주문내역 :", order)
    print("감사합니다 안녕히가세요")
    print("현재 재고:", inventory)
    break
    
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
inventory = {'팥붕어빵' : 10, '슈크림붕어빵' : 10, '초코붕어빵' : 10}

while True :
    print("\n 1. 팥붕어빵 : 1000원")
    print(" 2. 슈크림붕어빵 : 1200원")
    print(" 3. 초코붕어빵 : 1500원")
    print(" 4. 관리자모드")
    print(" 5. 종료")
    user_input = input("주문하실 붕어빵 번호를 입력해 주새요:")
    if user_input == "5" :
        print("프로그램을 종료 합니다.")
        break
    if user_input == '4':
        print("관리자 모드에 진입했습니다.")
        admin_input = input("추가할 붕어빵의 종류와 개수를 입력하세요 (예: 팥붕어빵5, 슈크림붕어빵3, 초코붕어빵2): ")
        
        # 재고 추가 처리
        for item in admin_input.split(","):  # 입력을 ','로 나눔
            kind = ''.join(filter(str.isalpha, item))  # 품목명 추출
            quantity = int(''.join(filter(str.isdigit, item)))  # 개수 추출
            if kind in inventory:
                inventory[kind] += quantity
                print(f"{kind}의 재고가 {quantity}개 추가되었습니다. 현재 재고: {inventory[kind]}")
            else:
                print(f"{kind}는 존재하지 않는 품목입니다.")
        print("재고 추가가 완료되었습니다.")
        continue
    
    order = {}
    if user_input == "1" :
        kind = "팥붕어빵"
    elif user_input == "2" :
        kind = "슈크림붕어빵"
    elif user_input == "3" :
        kind = "초코붕어빵"
    else :
        print("번호를 잘 못 입력하셨습니다. 다시 입력해 주세요")
        continue
    amount = int(input(f"{kind}를 선택하셨습니다. 수량을 입력해주세요:"))
    
    if kind in inventory :
        if inventory[kind] >= amount :
            inventory[kind] -= amount
            if kind in order :
                order[kind] += amount
            else : 
                order[kind] = amount
            print(f'{kind} {amount}개 판매완료')
        else :
            print(f"{kind}의 재고가 부족합니다. 최대 {inventory[kind]}만큼만 주문 가능합니다.")
            continue 
    else :
        print("판매하지 않는 상품입니다.")
        continue

    print("\n주문이 완료되었습니다.\n 주문내역 :", order)
    print("감사합니다 안녕히가세요")
    print("현재 재고:", inventory)
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# 초기 재고 및 가격 설정
inventory = {"팥붕어빵": 10, "슈크림붕어빵": 10, "초코붕어빵": 10}
prices = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}
sales = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}  # 판매된 개수
admin_password = "1234"  # 관리자 비밀번호 설정

while True:
    print("\n1. 팥붕어빵 (1000원)")
    print("2. 슈크림붕어빵 (1200원)")
    print("3. 초코붕어빵 (1500원)")
    print("4. 종료")
    print("5. 관리자 모드")
    user_input = input("주문할 붕어빵의 번호를 입력하세요: ")

    # 종료 조건
    if user_input == '4':
        print("프로그램을 종료합니다.")
        break

    # 관리자 모드
    if user_input == '5':
        password = input("관리자 비밀번호를 입력하세요: ")
        if password != admin_password:
            print("비밀번호가 틀렸습니다. 메인 화면으로 돌아갑니다.")
            continue

        print("\n관리자 모드에 진입했습니다.")
        while True:
            admin_input = input("1: 재고 추가, 2: 재고 조회, 3: 총 매출 조회, 4: 품목별 판매량 조회, 5: 관리자 모드 종료\n원하는 작업을 선택하세요: ")

            # 재고 추가
            if admin_input == "1":
                restock_input = input("추가할 붕어빵의 종류와 개수를 입력하세요 (예: 팥붕어빵5, 슈크림붕어빵3): ")
                for item in restock_input.split(","):  # 입력을 ','로 나눔
                    kind = ''.join(filter(str.isalpha, item))  # 품목명 추출
                    quantity = int(''.join(filter(str.isdigit, item)))  # 개수 추출
                    if kind in inventory:
                        inventory[kind] += quantity
                        print(f"{kind}의 재고가 {quantity}개 추가되었습니다. 현재 재고: {inventory[kind]}")
                    else:
                        print(f"{kind}는 존재하지 않는 품목입니다.")
                print("재고 추가가 완료되었습니다.")

            # 재고 조회
            elif admin_input == "2":
                print("\n현재 재고 상태:")
                for kind, count in inventory.items():
                    print(f"{kind}: {count}개")
                print()

            # 총 매출 조회
            elif admin_input == "3":
                total_revenue = sum(prices[kind] * sales[kind] for kind in sales)
                print("\n총 매출:")
                for kind, count in sales.items():
                    print(f"{kind}: {count}개 판매 (매출: {prices[kind] * count}원)")
                print(f"총 매출액: {total_revenue}원\n")

            # 품목별 판매량 조회
            elif admin_input == "4":
                print("\n품목별 판매량:")
                for kind, count in sales.items():
                    print(f"{kind}: {count}개 판매")
                print()

            # 관리자 모드 종료
            elif admin_input == "5":
                print("관리자 모드를 종료합니다.\n")
                break

            else:
                print("잘못된 입력입니다. 다시 선택해주세요.")

        continue

    # 주문 처리
    order = {}
    while True:
        if user_input == "1":
            kind = "팥붕어빵"
        elif user_input == "2":
            kind = "슈크림붕어빵"
        elif user_input == "3":
            kind = "초코붕어빵"
        else:
            print("잘못된 번호입니다. 다시 입력해주세요.")
            break

        quantity = int(input(f"{kind} 몇 개를 주문하시겠습니까? "))
        
        # 재고보다 많이 주문한 경우 처리
        if quantity > inventory[kind]:
            print(f"{kind}의 재고가 부족합니다. {inventory[kind]}개만 판매 가능합니다.")
            quantity = inventory[kind]

        if kind in order:
            order[kind] += quantity
        else:
            order[kind] = quantity

        another_order = input("추가로 주문하시겠습니까? (예/아니오): ")
        if another_order.lower() != "예":
            break

    print("\n주문 내역:", order)

    for kind, quantity in order.items():
        if kind in inventory:
            inventory[kind] -= quantity
            sales[kind] += quantity  # 판매 개수 누적
            print(f'{kind} {quantity}개 판매 완료')
        else:
            print(f'{kind}는 판매하지 않는 품목입니다.')

    print("\n현재 재고:", inventory)
```
