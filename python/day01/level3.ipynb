__[문제1]__

어떤 공원의 크기는 가로 50m, 세로 30m입니다.

공원 내부에 5개의 조형물이 있으며, 각 조형물의 면적은 10m²입니다.

이 공원의 실제 사용 면적을 구하세요. (cm로 구해주세요)




```python
park_area=50*30*10000
top=5
top_area=top*10*10000
true_area=park_area-top_area
print(true_area)
#1m**2=10000cm**
```

__[문제2]__

2권씩 묶여서 팔리는 책의 가격은 12,000원입니다.

나는 3묶음을 살 겁니다.

근데 최종 가격을 보니 10%의 세금이 붙어있습니다.

최종 가격을 구해주세요.

```python
two_book=12000
buy_book=3
tex=0.1
total_tex=two_book*buy_book*tex
print(two_book*buy_book+total_tex)
```

__[문제3]__

가로 20m, 세로 15m인 직사각형 정원의 면적을 구하세요.


```python
width=20
length=15
area=width*length
print(area)
```

__[문제3-1]__

문제 1에서 구한 정원의 면적에서, 정원의 5%를 차지하는 풀밭 면적을 구하세요.




```python
grass_area_percent=0.05
grass_area = area*grass_area_percent
print(grass_area)
```
