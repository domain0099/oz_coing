```python
class Person :
    def __init__(self, name, age, address):
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print("{0} 저는 {1} 입니다.".format(self.hello, self.name))
maria = Person('마리아', 20, '서울시 서초구 반포동')
maria.greeting() #안녕하세요. 저는 마리아입니다.

print('이름', maria.name)    #마리아
print('나이', maria.age)     #20
print('주소', maria.address) # 서울시 서초구 반포동
```

```python
class Car(object):
    maxspeed = 300
    maxpeople = 5
    def move(self, x):
        print('스피드로 달리고 있습니다')
    def stop(self):
        print('멈췄습니다')

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

class electricCar(Car):
    battery = 1000
    batteryKM = 300
    def move(self, x):
        print(self.batteryKM, '만큼 달릴 수 있습니다')
        print(x, '스피드로 달리고 있습니다.')

k5 = HybridCar()
k5.maxspeed

electricCark5 = electricCar()
electricCark5.maxspeed
electricCark5.battery
electricCark5.move(10)
#print(dir(k5))
```

```python
#예외코드작성

try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    print(y)
except :
    print('예외가 발생했습니다')
```

```python
y = [10, 20, 30]
try :
    index, x = map(int, input('나눌 숫자를 입력하세요: ').split(','))
    print(y[index] /x)
except :
    print('오류입니다')
```

```python
y = [10, 20, 30]
try :
    index, x = map(int, input('나눌 숫자를 입력하세요: ').split(','))
    print(y[index] /x)
except ZeroDivisionError : #숫자를 0으로 나눴을때 발생하는 에러를 처리하는 기능
    print('숫자를 0으로 나눌 수 없습니다.')
except IndexError : # 잘못된 인덱스를 입력했을때 발생하는 에러를 처리하는 기능
    print('잘못된 인덱스를 입력했습니다')
```

```python
try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    print(y)
except :
    print('예외가 발생했습니다')
else :
    print(y, '성공하셨습니다') #여기서 else는 정상적으로 작성하였을때 코드를 작성하는 공간
```

```python
try:
    x = int(input('나눌 숫자를 입력하세요'))
    y = 10 / x
    print(y)
except :
    print('예외가 발생했습니다')
else :
    print(y, '성공하셨습니다')  # finally는 무조건 실행이된다
finally :
    print("코드 실행이 종료 되었습니다.")
```
