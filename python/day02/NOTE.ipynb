```python
숫자 = '1234566778998675856'
#전체출력
print(숫자[::])

#거꾸로 출력
print(숫자[::-1])

#짝수출력
print(숫자[::2])

#홀수출력
print(숫자[1::2])


print(숫자[4:1:-1])
print(숫자[1:4:])

```
